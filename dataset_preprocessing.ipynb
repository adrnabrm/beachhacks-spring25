````python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("paultimothymooney/recipenlg")

print("Path to dataset files:", path)
````

````python
import os
import pandas as pd

csv_path = os.path.join(path, "RecipeNLG_dataset.csv")
df = pd.read_csv(csv_path)
````

````python
df.head()
````

````python
import numpy as np
import ast
DATASET_SIZE = 100000
SYSTEM_PROMPT = """You are an AI chef. Given a list of ingredients, generate a recipe in the following JSON format that will always have a title, ingredients WITH specified measurements, and detailed instructions:
  ```json
{
  "title": "Recipe title",
  "ingredients": ["ingredient1", "ingredient2", ...],
  "directions": ["step1", "step2", ...]
}
"""
dataset = []
# Example format of each piece of training data
"""
{
  'system': some message,
  'input': [ingredients],
  'output: {
    'title': some title,
    'ingredients': [some ingredients of specified measurements],
    'directions': [some directions]
    }
}"""
for i in range(DATASET_SIZE):
  try:
    random_row = df.iloc[np.random.randint(df.shape[0])]

    user_input = ast.literal_eval(random_row['NER'])

    title = random_row['title']
    ingredients = ast.literal_eval(random_row['ingredients'])
    directions = ast.literal_eval(random_row['directions'])
    dataset.append({
      'system': SYSTEM_PROMPT,
      'input': user_input,
      'output': {
        'title': title,
        'ingredients': ingredients,
        'directions': directions
      }
    })
  except:
    continue
````

````python
import json

with(open('dataset.json', 'w')) as f:
  json.dump(dataset, f, indent=4)
````
